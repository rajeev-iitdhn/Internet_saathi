```
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.utils import to_categorical
from keras.preprocessing import image
from sklearn.preprocessing import LabelEncoder
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
from tqdm import tqdm_notebook
from skimage import io
import urllib.request
```

```
train = pd.read_csv('/content/drive/My Drive/MachineLearning/gender_results/improve/train_gender_for_3000.csv')
```

```
len(train)
```

```
encoder = LabelEncoder()
encoder
```

```
train_image = []
for i in tqdm_notebook(range(train.shape[0])):
  img = image.load_img('/content/drive/My Drive/MachineLearning/gender_results/improve/train_3000/'+train['image_name'][i], target_size=(28,28,3), grayscale=False)
  img = image.img_to_array(img)
  img = img/255
  train_image.append(img)
```

```
len(train_image)
```

```
X = np.array(train_image)
```

```
y = train['Gender'].values
```

```
encoder.fit(y)
```

```
y = encoder.transform(y)
y
```

```
y = to_categorical(y)
```

```
y.shape
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

```
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),activation='relu',input_shape=(28,28,3)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(2, activation='softmax'))
```

```
model.compile(loss='categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```
model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))
```
