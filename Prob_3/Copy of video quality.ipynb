```
df=pd.read_csv('video_urls.csv')
```

```
df.head()
```

```
links = list(df['links'])
links
```

```
import cv2

res_list = []
quality_list = []
for i in range(len(links)):
  cap = cv2.VideoCapture(links[i])

  height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
  width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
  if (height<720.0):
    quality_list.append("Low")
  else:
    quality_list.append('High')
  res_list.append((width, height))
```

```
df['Resolution'] = res_list
df['Quality'] = quality_list
df.head(20)
```
