```
from PIL import Image 
import matplotlib.pyplot as plt 
import numpy as np
import face_recognition
import keras 
from keras.models import load_model
import cv2
```

### Emotion detection

#### trial1

```
emotion_dict = {'Angry':0, 'Sad':5, 'Neutral':4, 'Disgust': 1, 'Surprise':6, 'Fear':2, 'Happy':3}
```

```
img = face_recognition.load_image_file('/content/drive/My Drive/MachineLearning/faces/test_images/9a865d4ad8cfbe662660032fc58cd64e.jpg')
plt.imshow(img)
```

```
face_locations = face_recognition.face_locations(img)
face_locations
```

```
top, right, bottom, left = face_locations[0]
face_img = img[top:bottom, left:right]
plt.imshow(face_img)
```

```

face_img = cv2.resize(face_img, (48,48))
face_img = cv2.cvtColor(face_img, cv2.COLOR_BGR2GRAY)
face_img = np.reshape(face_img, [1, face_img.shape[0], face_img.shape[1], 1])
```

```
# Load the model 
model = load_model('/content/drive/My Drive/MachineLearning/faces/model_v6_23.hdf5')
```

```
face_img.shape
```

```
predict_class = np.argmax(model.predict(face_img))
label_map = dict((v,k) for k,v in emotion_dict.items())
predict_label = label_map[predict_class]
```

```
predict_label
```
