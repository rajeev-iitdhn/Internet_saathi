**Taking saathis who has the assessment score.**

```
table=data[data.percent_ass_score.notna()]
```

**Below table shows the impact of all the parameters on each other.**  
**Correlation coefficient is not the exact slope of line between two variables. But it is a very good reference parameter which shows impact of one variable on other.**

```
corr=table.corr(method="spearman").percent_ass_score
```

```
corr
```

**perf_score is created as a general performance assessment parameter for all the present saathis. It is by using the impact of input variables on given assessment score.**

```
original_data["perf_score"]=(corr.NoOfFormsWorked*original_data.NoOfFormsWorked)+(corr.percent_target_achieved*original_data.percent_target_achieved)+(corr.NoOfBeneficiaryToBeFilled*original_data.NoOfBeneficiaryToBeFilled)+(corr.BeneficiariesFilled*original_data.BeneficiariesFilled)+(corr.vill_count*original_data.vill_count)+(corr.block_count*original_data.block_count)+(corr.OnFieldDaysCount*original_data.OnFieldDaysCount)+(corr.SystemDaysCount*original_data.SystemDaysCount)-(corr.meanTimePerQuestion*original_data.meanTimePerQuestion)-(corr.engagedWithLivelihoodActivity*original_data.engagedWithLivelihoodActivity)+(corr.numberOfChildren*original_data.numberOfChildren)+(corr.educationalQualification*original_data.educationalQualification)+(corr.age*original_data.age)+(corr.haveAadhaarCard*original_data.haveAadhaarCard)+(corr.haveAadhaarCard*original_data.haveAtmCard)+(corr.percent_target_achieved*original_data.percent_target_achieved)
```
